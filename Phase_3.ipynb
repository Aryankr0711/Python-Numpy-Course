## NUMPY on real world dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([ # It a 2D Array
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print('=='*5, 'Zomato sales Analysis', '=='*5)
print('sales data Shape is: ', sales_data.shape)
print('\n Original data is:\n')
#print(sales_data[:5:])
print(sales_data[:, 1:]) # see the shape its 5x5 so there are 5 rows and 5 columns
```

## Total sales per year

```python
print('The Total sales for Paradise Biryani 21-24 is: ', sum(sales_data[0, 1:]))
print('The Total sales on Zomato 23 is: ', np.sum(sales_data[ : , 1:], axis=0))

# if wanted to find the sum of all rows at once, use axis as show above
# same as the previous one, axis=0, denotes columns and axis=1 denotes rows
```

```python
# Print sing loops slaes for each year

for i in range (1,5,1):
  print('The Total sales on Zomato',20+i, 'is: ', sum(sales_data[ : , i]))
```

```python
# maximum sales per resturant
max_sale= np.sum(np.max(sales_data[:, 1: ], axis=0))
print(max_sale)
print('\n')
min_sale= np.sum(np.min(sales_data[:, 1: ], axis=0))
print(min_sale)
```

```python
# Average sales per resuturat
avg_sale= np.mean(sales_data[:, 1:], axis=1)# as row wise
print(avg_sale)
print('Avg is: ', avg_sale)
```

```python
# cumulative sales
cumsum=np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)
```

```python
vector=np.array([1, 2, 3, 4])
vector2=np.array([8, 7, 6, 5])

print('Vector Addition: ', vector+vector2)
print('Vector Multiplication: ', vector*vector2)
print('Vector Dont product: ', np.dot(vector, vector2))
```

```python
 # vectorized Operations
resturent_types=np.array(['burger', 'pizza','coffee', 'food'])

vectorized_upper=np.vectorize(str.upper)
print(vectorized_upper(resturent_types))
 # vectorized Operations are perfored to each element one by one
 # What so ever is the opertion
```

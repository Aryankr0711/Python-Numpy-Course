# Why Numpy over loops?

```python
arr1= np.random.random(10000000)*100
arr2= np.random.random(10000000)*100
start_time= time.time()
count=0;

for i in range(len(arr1)): # Iterate over indices
    count+=arr1[i]*arr2[i] # Use 'i' as index
end_time=time.time()

print("Time taken is: ", end_time-start_time)
```

```python
arr1= np.random.random(10000000)*100
arr2= np.random.random(10000000)*100
start_time= time.time()
arr3=np.dot(arr1, arr2)
end_time=time.time()

print("Time taken is: ", end_time-start_time)
```

## Creates and Array using inbuilt functions

```python
#  Zeroes
zero=np.zeros((4, 4))
print("Array of Zeros is: \n", zero)

print('\n\n')
ones=np.ones((4, 4))
print("Array of Ones is: \n", ones)
print('\n\n')
full=np.full((4, 4), 7)
print("Array of 7 is: \n", full)
print('\n\n')
random=np.random.random((4, 4))
print("Array of Random Numbers are: \n", random)
print('\n\n')
sequence=np.arange(10, -1, -1)
print("Array of sequence are: \n", sequence)
```

## Array Properties

```python
arr= np.array([[1, 2, 3], [4, 5, 6]])

#shape
print('shape is: ', arr.shape)
print('Dimension is: ', arr.ndim)
print('Size is: ', arr.size)
print('Data type is: ', arr.dtype)
```

```python
arr=np.arange(12)

print('original Array is: ', arr)
print('\n\n')

reshaped= arr.reshape(3, 4)
print('Reshaped is: ', reshaped)
print('\n\n')

falt= reshaped.flatten()
print('Flaterened is: ', falt)
print('\n\n')

ravel= reshaped.ravel()
print('Ravelled is: ', ravel)
print('\n\n')
# flatten creates a copy whereas ravel creates a view

transpose=reshaped.T
print('Transposed is: ', transpose)
print('\n\n')
```

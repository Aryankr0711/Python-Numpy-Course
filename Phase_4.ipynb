```python
import numpy as np
```

```python
array1=np.array([[1, 2, 3], [4, 5, 6]])
array2=np.random.rand(3, 3)
array3=np.ones((4, 4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
load_array1=np.load('array1.npy')
print(load_array1)
print('\n')
load_array1=np.load('array2.npy')
print(load_array1)
print('\n')
load_array1=np.load('array3.npy')
print(load_array1)
```
